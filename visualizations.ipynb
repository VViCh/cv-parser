# CV Parser NER Evaluation Visualizations
This notebook contains the code to generate the training metric comparisons and confusion matrices.

```python
import json
import matplotlib.pyplot as plt
import numpy as np

# Load data
with open('models/training_results.vfinal-final.v2.json', 'r') as f:
    data = json.load(f)

models = list(data.keys())
f1_scores = [data[m]['best_f1'] for m in models]
precisions = [data[m]['best_precision'] for m in models]
recalls = [data[m]['best_recall'] for m in models]

sorted_indices = np.argsort(f1_scores)[::-1]
models = [models[i] for i in sorted_indices]
f1_scores = [f1_scores[i] for i in sorted_indices]
display_names = [m.split('/')[-1] for m in models]

plt.figure(figsize=(12, 6))
bars = plt.bar(display_names, f1_scores, color='skyblue')
plt.title('Best F1 Score by Model', fontsize=16)
plt.ylabel('F1 Score', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, max(f1_scores) + 0.1)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, round(yval, 3), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
